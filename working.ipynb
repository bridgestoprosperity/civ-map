```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Point
```

```python
# read file as geodataframe
gdf = gpd.read_file('/Users/camkruse/Downloads/civ-assessments-v1.csv')
```

```python
gdf
```

```python
# return row with latitude matching ''
gdf.loc[gdf['longitude'] == '']
```

```python
# convert latitude to float
gdf['latitude'] = gdf['latitude'].astype(float)
gdf['longitude'] = gdf['longitude'].astype(float)
```

```python
# set geometry to latitude and longitude columns
gdf['geometry'] = gdf.apply(lambda x: Point((float(x.longitude), float(x.latitude))), axis=1)
```

```python
# save to GeoJSON
gdf.to_file('/Users/camkruse/Downloads/civ-assessments-v1.geojson', driver='GeoJSON')
```
