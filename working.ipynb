```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Point
import unicodedata
```

```python
# read file as geodataframe
gdf = gpd.read_file('./home/data/civ-assessments-v1.csv', encoding='iso-8859-1')
gdf
```

```python
# return row with latitude matching ''
gdf.loc[gdf['longitude'] == '']
```

```python
# convert latitude to float
gdf['latitude'] = gdf['latitude'].astype(float)
gdf['longitude'] = gdf['longitude'].astype(float)
```

```python
# set geometry to latitude and longitude columns
gdf['geometry'] = gdf.apply(lambda x: Point((float(x.longitude), float(x.latitude))), axis=1)
```

```python
# save to GeoJSON
gdf.to_file('./home/data/civ-assessments-v5.geojson', driver='GeoJSON')
```

```python
# save to csv
gdf.to_csv('./home/data/civ-assessments-v5.csv')
```

```python
# read './data/civ-assessments-v5.geojson' ad gdf
filtergdf = gpd.read_file('./home/data/civ-assessments-v5.geojson', encoding='iso-8859-1')
filtergdf
```

```python
# convert "River crossing deaths in last 3 years" to int
filtergdf['River crossing deaths in last 3 years'] = filtergdf['River crossing deaths in last 3 years'].astype(int)
```

```python
# create a dataframe with rows where "River crossing deaths in last 3 years" is greater than 0
mgdf = filtergdf[filtergdf['River crossing deaths in last 3 years'] > 0]
mgdf
```

```python
# create a dataframe with rows where "Flag for Rejection" No
ngdf = filtergdf[filtergdf['Flag for Rejection'] == 'No']
ngdf
```

```python
# create a dataframe from mgdf where "Flag for Rejection" is No
mngdf = mgdf[mgdf['Flag for Rejection'] == 'No']
mngdf
```

```python
# write mgdf, ngdf, and mn gdf to geojson files
mgdf.to_file('./home/data/civ-assessments-v5-mgdf.geojson', driver='GeoJSON')
ngdf.to_file('./home/data/civ-assessments-v5-ngdf.geojson', driver='GeoJSON')
mngdf.to_file('./home/data/civ-assessments-v5-mngdf.geojson', driver='GeoJSON')
```
